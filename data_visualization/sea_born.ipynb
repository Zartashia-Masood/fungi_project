```python
#strip plot 
#used to display the distribution of individual data points along a categorical axis,
#Show raw data points,
#Visualize the spread of values for each category,
#Detect clusters, outliers, and data density.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the planets dataset
planets = sns.load_dataset('planets')

# Drop rows with missing values in 'distance'
planets = planets.dropna(subset=['distance'])

# Subset: only include planets within 100 light years
subset = planets[planets['distance'] <= 100]
#Keeps only the planets that are â‰¤ 100 light years away.

# Create a strip plot showing method vs distance
plt.figure(figsize=(12, 6))
sns.stripplot(data=subset, x='method', y='distance', jitter=True)
#jitter=True: Adds a small random "spread" horizontally to avoid overlapping dots.

plt.title('Exoplanet Detection Methods for Planets Within 100 Light Years')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#swarm plot 
#Plots each individual data point,
#Automatically adjusts positions of points so they donâ€™t overlap (unlike strip plots),
#Shows data density, spread, and individual values clearly.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
planets = sns.load_dataset('planets')

# Drop missing values in 'distance' to avoid plotting issues
planets = planets.dropna(subset=['distance'])

# Subset the data: only planets within 100 light years
subset = planets[planets['distance'] <= 100]

# Set up the plot
plt.figure(figsize=(12, 6))
sns.swarmplot(data=subset, x='method', y='distance')

# Customize plot
plt.title('Exoplanet Detection Methods for Planets Within 100 Light Years (Swarm Plot)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#Each dot = one exoplanet.

```

```python
#histogram
#It groups data into ranges (bins),
#The result is shown as bars:
#Height of the bar = frequency (count)
#X-axis = value ranges
#Y-axis = number of occurrences (how many times)

import seaborn as sns
import matplotlib.pyplot as plt

# Load the planets dataset
df = sns.load_dataset("planets")

# Drop missing values in 'mass'
df_clean = df.dropna(subset=["mass"])

# Plot histogram of planet masses
plt.figure(figsize=(8, 5))
sns.histplot(data=df_clean, x="mass", bins=30, kde=True, color="skyblue")
#kde=True:adds a Kernel Density Estimate (KDE) curve â€” a smooth curve that estimates the distribution
#bins=30:Divides the data into 30 equal-width bins (bars in the histogram)

# Labels and title
plt.title("Distribution of Exoplanet Masses")
plt.xlabel("Mass (Jupiter masses)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#regression plot 
#Scatter plot â€” shows individual data points,
#regression line â€” a best-fit straight line through the data, showing the trend or correlation
#Confidence interval (optional) â€” shaded area around the line showing uncertainty.

import seaborn as sns
import matplotlib.pyplot as plt

# Drop missing values for mass and distance
df_clean = df.dropna(subset=["mass", "distance"])

# Create regression plot
plt.figure(figsize=(8, 6))
sns.regplot(data=df_clean, x="distance", y="mass", scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})

#scatter_kws allows you to customize the scatter plot dots.
#"alpha": 0.5 sets transparency to 50%, so overlapping dots become more visible.

#line_kws customizes the regression line.
#"color": "red" draws the regression (trend) line in red.

# Labels and title
plt.title("Regression: Planet Mass vs Distance")
plt.xlabel("Distance (parsecs)")
plt.ylabel("Mass (Jupiter masses)")
plt.tight_layout()
plt.show()
```

```python
#line plot
#A line plot (or line chart):
#Plots data points on a 2D graph.
#Connects them with lines to show trends over a continuous variable, typically time.
#Useful for tracking changes or patterns over intervals.

sns.set_context("notebook")
sns.set_style("whitegrid")

import seaborn as sns

# Plot lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(data=planets, x="year", y="number", color="purple")

#Add titles and labels
plt.title('Number of Planets Discovered Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Planets')

plt.show()
```

```python
#joint plot
#A joint plot combines two plots to visualize the relationship between two variables:
#Scatter plot (or other plot type) in the center, 
# showing the relationship between two variables (X and Y)
#Histograms (or density plots) on the top and right margins,
# showing the distribution of each variable individually.

sns.set_context("notebook")
sns.set_style("whitegrid")

import seaborn as sns
import matplotlib.pyplot as plt


# Create a joint plot
g= sns.jointplot(data=planets, x="distance", y="mass", kind="scatter", color="skyblue",alpha=0.5)

#kind='scatter' creates a scatter plot with marginal histograms.
#alpha=0.5 makes points semi-transparent for bette
# Adjust the title to avoid overlap with the plot

# Add title to the figure (not just one axis)
g.fig.suptitle('Distance vs Mass of Planets', fontsize=16)

# Adjust the title to avoid overlap with the plot
g.fig.tight_layout()


plt.show()
```

```python
#bar plot
sns.set_context("notebook")
sns.set_style("whitegrid")
#similar to histogram
sns.barplot(data=planets,x="method",y="mass",palette="Set2")

# Add titles and rotate x labels for readability
plt.title('Average Mass of Planets by Discovery Method')
plt.xlabel('Discovery Method')
plt.ylabel('Average Mass')
plt.xticks(rotation=45)
#xticks:rotates the method names so they donâ€™t overlap.
plt.show()
```

```python
#count plot
#Plots counts of observations for each category.
#Useful for categorical variables.
#Shows the distribution of categories.


# Plot count plot for 'method'
plt.figure(figsize=(12, 6))
sns.countplot(data=planets, x='method', palette='pastel')
# add title
plt.title('Count of Planets Discovered by Method')
plt.xlabel('Discovery Method')
plt.ylabel('Number of Planets')
plt.xticks(rotation=45)  # Rotate labels for better readability

plt.show()
```

```python
#box plot
#A box plot (also called a box-and-whisker plot) is a graphical summary of a dataset
# that shows its central value, spread, and outliers.
#The median
#The interquartile range (IQR)
#The minimum and maximum (excluding outliers)
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("planets")

# Drop missing values
df_clean = df.dropna(subset=["mass"])

# Create box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_clean, x="method", y="mass", palette="Set2")

# Customize
plt.xticks(rotation=90)
plt.title("Distribution of Planet Mass by Discovery Method")
plt.xlabel("Discovery Method")
plt.ylabel("Mass (Jupiter masses)")
plt.tight_layout()
plt.show()
```

```python
#violin plot

#A violin plot is a combination of a:

#boxplot (shows median, quartiles, and range), and

#KDE plot (kernel density estimation) â€” showing the distribution shape of the data.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset("planets")

# Drop missing values for 'mass'
df_clean = df.dropna(subset=["mass"])

# Create the violin plot
plt.figure(figsize=(12, 6))
#Width = 12 units: useful for plotting many discovery methods along the x-axis.

#Height = 6 units: gives enough room to show mass variation along the y-axis.
sns.violinplot(data=df_clean, x="method", y="mass", palette="Set3", inner="box")

# Customize the plot
plt.xticks(rotation=90)
plt.title("Distribution of Planet Mass by Discovery Method")
plt.xlabel("Discovery Method")
plt.ylabel("Mass (Jupiter masses)")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("planets")

# Drop missing values
df_clean = df.dropna(subset=["mass"])
#This step is crucial because rug plot gives error for NaN 

# Create rug plot
plt.figure(figsize=(8, 2))
#Initializes a new plot figure using matplotlib.

#figsize=(8, 2) creates a wide and short horizontal canvas.

#This is a common shape for rug plots since they are usually compact and emphasize density along one axis.
sns.rugplot(data=df_clean, x="mass", height=0.1, color="darkblue")

plt.title("Rug Plot: Planet Mass Distribution")
plt.xlabel("Mass (Jupiter masses)")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#seaborn is used for statistical data visualization.

#matplotlib.pyplot is used for plotting control and layout customization.

# Load dataset
df = sns.load_dataset("planets")

# Select numeric columns and drop NaNs
df_numeric = df[["number", "mass", "distance", "orbital_period"]].dropna()

#Selects only numerical columns relevant for analysis.

#Drops any rows with missing values (NaN) to avoid issues when plotting.

# Create pair plot
sns.pairplot(data=df_numeric, corner=True, diag_kind="kde", plot_kws={"alpha": 0.6})

#sns.pairplot(...):

#Creates a grid of pairwise plots for all combinations of the selected numerical variables.

#Useful to visualize correlations and distributions.

#data=df_numeric: Uses the cleaned numeric DataFrame.

#corner=True:

#only plots the lower triangle of the pairwise matrix.

#Avoids redundant mirrored plots from the upper triangle.

#Makes the plot less cluttered and easier to read.

#diag_kind="kde":

#Plots kernel density estimates (KDEs) on the diagonal.

#KDE gives a smoothed version of the histogram (shows distribution of individual variables).

#plot_kws={"alpha": 0.6}:

#Sets transparency of scatterplot markers (helps when points overlap).

#alpha=0.6 means 60% opacity.

plt.suptitle("Pair Plot of Planet Properties", y=1.02)
#Adds a main title to the entire figure.

#y=1.02 moves the title slightly above the plot to avoid overlapping with axes.
plt.tight_layout()
#Adjusts subplot spacing to prevent overlap between axes, labels, and title.
plt.show()
#show final result 
#Scatterplots between each pair of:

#number, mass, distance, orbital_period

#Diagonal plots show distribution of each variable using KDE.

#helps to visually identify:

#correlations (linear or nonlinear relationships)

#Clusters or outliers

#Skewness in distributions
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the planets dataset
planets = sns.load_dataset("planets")

# Drop missing values for simplicity
planets_clean = planets.dropna()

# Create a PairGrid
g = sns.PairGrid(planets_clean, vars=["mass", "orbital_period", "distance"], hue="method")

# Map plots to the grid
g.map_diag(sns.histplot)               # Histogram on the diagonal
g.map_offdiag(sns.scatterplot)         # Scatterplots off the diagonal

# Add legend
g.add_legend()
#vars=["mass", "orbital_period", "distance"]: Specifies the numerical columns to visualize.

#hue="method": Colors the data points by the planet detection method.

#map_diag: Maps a histogram to the diagonal plots.

#map_offdiag: Maps scatter plots to the off-diagonal cells.

# Show plot
plt.show() 
```
