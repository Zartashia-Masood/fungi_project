```python
# stem height 
import matplotlib.pyplot as plt
plt.hist(df["stem-height"],bins=30,edgecolor='black')
#add title 
plt.title("heights of mushroom's stem")
plt.xlabel("stem-height")
plt.ylabel("frequency")
plt.show()
```

```python
#stem width
plt.hist(df["stem-width"],bins=35,edgecolor='black')

#add title 
plt.title("width of mushroom's width")
plt.xlabel("stem-width")
plt.ylabel("frequency")
plt.show()
```

```python
#boxplot
plt.boxplot(df["cap-diameter"])
#add title
plt.title("diameter of mushroom cap")
plt.xlabel("cap-diameter")
plt.ylabel("frequency")
plt.show()
```

```python
#distribution of cap shapes 
cap_shape_counts = df['cap-shape'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(cap_shape_counts, labels=cap_shape_counts.index, autopct='%1.1f%%', startangle=90)
#autopct='%1.1f%%'
#This formats the percentage labels on the pie slices.
#'%1.1f%%' means:Show percentages with 1 decimal place
#startangle=90:Rotates the chart so it starts at 90 degrees (the top).

#add title 
plt.title("Distribution of Cap Shapes")
plt.tight_layout()
plt.show()
```

```python
#gill color distribution
gill_color_counts = df['gill-color'].value_counts()

plt.figure(figsize=(10, 4))
plt.bar(gill_color_counts.index, gill_color_counts.values, color="purple")

#add title 
plt.title("Gill Color Distribution")
plt.xlabel("Gill Color")
plt.ylabel("Count")
plt.xticks(rotation=45) #ease in readabilty by rotating 
plt.tight_layout()
plt.show()
```

```python
#counts of class

class_counts = df['class'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(class_counts.index, class_counts.values, color=["green", "red"])

#add title 
plt.title("Count of Edible vs Poisonous Mushrooms")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=["Edible", "Poisonous"])
plt.tight_layout()
plt.show()
```

```python
# cap colors top 5
cap_colors = df['cap-color'].value_counts().head(5)

plt.figure(figsize=(6, 4))
plt.bar(cap_colors.index, cap_colors.values, color='orange')

#add title
plt.title("Top 5 Cap Colors")
plt.xlabel("Cap Color")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#cap color by mushroom class
cap_color_class = pd.crosstab(df['cap-color'], df['class'])

cap_color_class.plot(kind='bar', figsize=(10, 5), color=['green', 'red'])

#add title
plt.title("Cap Color by Mushroom Class")
plt.xlabel("Cap Color")
plt.ylabel("Count")
plt.legend(["Edible", "Poisonous"]) 
#legend:a box or label area in a plot that explains
#what the colors, symbols, or patterns represent
plt.xticks(rotation=45) #the small marks along the horizontal axis that indicate values or categories.
plt.tight_layout()
plt.show()
```

```python
habitat_counts = df['habitat'].value_counts().sort_values()

plt.plot(habitat_counts.index, habitat_counts.values, marker='o', linestyle='-', color='teal')
#add title
plt.title('Habitat Frequency')
plt.xlabel('Habitat')
plt.ylabel('Count')
plt.grid(True) #grid
plt.show()
```

```python
cross_tab = pd.crosstab(df['cap-color'], df['class'])

cross_tab.plot(kind='bar', stacked=True, color=['blue', 'red'])
plt.title('Cap Color by Class')
plt.xlabel('Cap Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Class')
plt.tight_layout()
plt.show()
```
